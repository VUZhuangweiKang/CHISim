```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
term_policies = ['Random', 'LeastCoreAssigned', 'LeastResubmit', 'MostRecentDeployed']
```

# Completed Jobs

```python
baseline_completed_jobs = np.array([60044, 45744, 58559, 57383])
rolling_mean_completed_jobs = np.array([54984, 42933, 53753, 53560])
rolling_median_completed_jobs = np.array([56624, 42698, 53876, 54534])
lstm_completed_jobs = np.array([55692, 43532, 53240, 53652])
```

```python
x = np.arange(4)
total_width, n = 1, len(term_policies)
width = total_width / n

plt.xticks(x, term_policies, rotation=15)
x = x - (total_width - width)/1.5

# plt.bar(x, baseline_completed_jobs/baseline_completed_jobs, label='Baseline', width=width)
plt.bar(x+width, rolling_mean_completed_jobs/baseline_completed_jobs, label='RollingAvg', width=width)
plt.bar(x+2*width, rolling_median_completed_jobs/baseline_completed_jobs, label='RollingMedian', width=width)
plt.bar(x+3*width, lstm_completed_jobs/baseline_completed_jobs, label='LSTM', width=width)


plt.ylim(top=1, bottom=0.8)
plt.legend(bbox_to_anchor=(0.5, 1.1), loc='center', ncol=3)
plt.xlabel('Node Termination Policy')
plt.ylabel('Completed Jobs Relative to Baseline')
plt.rc('axes', axisbelow=True)
plt.grid(axis='x', linestyle=':')
plt.savefig('completed_jobs.png', dpi=500, bbox_inches='tight')
```
