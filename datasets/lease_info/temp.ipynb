```python
import pandas as pd
from datetime import datetime
```

```python
get_timestamp = lambda time_str: datetime.strptime(time_str, '%Y-%m-%d %H:%M:%S').timestamp()
df = pd.read_csv('compute_haswell.csv')
df['start_on'] = pd.to_datetime(df['start_on'])
df['created_at'] = pd.to_datetime(df['created_at'])
df['end_on'] = pd.to_datetime(df['end_on'])
df['deleted_at'] = pd.to_datetime(df['deleted_at'])
df['duration'] = df['end_on'] - df['start_on']
time_diff = (df['start_on'] - df['created_at']).to_list()
time_diff = pd.DataFrame([x.total_seconds() for x in time_diff])
dur = df['duration'].to_list()
dur = pd.DataFrame([x.total_seconds() for x in dur])
dur.plot()
```
