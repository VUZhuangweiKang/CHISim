```python
import pandas as pd
```

```python
df = pd.read_csv('lease_info-bk.csv')
df = df.iloc[20000:]
temp = df[['node_cnt', 'trace_time']]
temp['trace_time'] = pd.to_datetime(temp['trace_time'])
temp.set_index(['trace_time'], inplace=True)
temp.resample('3H').sum().fillna(0).plot()
```
